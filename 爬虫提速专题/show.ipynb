并发：同一时间段内 交替执行 多个任务（宏观同时，微观交替）
并发:  任务A | 任务B | 任务A | 任务B | 任务A
       |_____|     |_____|     |_____|
       CPU切换      CPU切换     CPU切换

并行：同一时刻 真正同时执行 多个任务
并行:  任务A | 任务A | 任务A | 任务A | 任务A
       任务B | 任务B | 任务B | 任务B | 任务B
       任务C | 任务C | 任务C | 任务C | 任务C
       CPU1  CPU2  CPU3  (多核同时执行)

多线程：在一个进程内创建多个 线程 ，每个线程独立执行
进程包含:
├── 主线程
├── 子线程1 ──→ 任务A
├── 子线程2 ──→ 任务B
└── 子线程3 ──→ 任务C

异步：不等待结果返回，先执行后续代码，结果通过 回调/Promise/async await 通知
同步：等待结果返回，后续代码执行会阻塞，直到结果返回
